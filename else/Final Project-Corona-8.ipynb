**We noticed that we have ordinal data representing dates in the `Counties` dataframe; we need to convert them into actual dates.** 

```python
def convert_ordinal(mydataframe, col):
    mydataframe[col] = mydataframe[col].fillna(0).astype('int64')
    mydataframe = mydataframe[mydataframe[col] != 0]
    mydataframe[col] = mydataframe[col].apply(dt.datetime.fromordinal)
    return mydataframe

#Counties1 = Counties[Counties['>500 gatherings']!=0]
Counties0 = convert_ordinal(Counties, '>500 gatherings')
Counties1 = convert_ordinal(Counties0, 'public schools')
Counties2 = convert_ordinal(Counties1, 'restaurant dine-in')
Counties3 = convert_ordinal(Counties2, 'entertainment/gym')
Counties4 = convert_ordinal(Counties3, 'federal guidelines')
Counties5 = convert_ordinal(Counties4, 'foreign travel ban')

Counties5
```

**In the data frame `Confirmed`, we noticed we have data on the number of confirmed cases throughout U.S. We want to creat a new dataframe to show the exact numbers from each County and State. Let's name the new dataframe `Confirmed2`**.  

```python
Confirmed0 = Confirmed.rename(columns={'Long_':'Long', 'Admin2':'County', 'Province_State':'Province/State', 'Country_Region':'Country/Region'})
Confirmed1 = Confirmed0.groupby(['Country/Region', 'Province/State', 'FIPS', 'County']).mean()
Confirmed2 = Confirmed1.drop('code3', axis=1)
Confirmed2
```

**The same treatment goes to dataframe `Death`, and let's name the new dataframe `Death2`.**

```python
Death0 = Death.rename(columns={'Long_':'Long', 'Admin2':'County', 'Province_State':'Province/State', 'Country_Region':'Country/Region'})
Death1 = Death0.groupby(['Country/Region', 'Province/State', 'FIPS', 'County']).mean()
Death2 = Death1.drop('code3', axis=1)
Death2
```

**From the dataframes `Confirmed2`, let's extract all the data for California and add a column `Confirmed Sum` for the sum of the confirmed cases for each county. Let's name the new dataframe `California_Confirmed_Sum`.**

Notice: We dropped two rows of invalid data from our dataframe.

```python
California_confirmed = Confirmed2.iloc[Confirmed2.index.get_level_values('Province/State') == 'California']
California_confirmed1 = California_confirmed[California_confirmed['Lat'] != 0]

Get_Confirmed_Sum = pd.DataFrame(California_confirmed1.drop(['UID', 'Lat', 'Long'], axis = 1).sum(axis = 1)).rename(columns = {0:'Comfirmed Sum'})

California_Confirmed_Sum = pd.concat([California_confirmed1, Get_Confirmed_Sum], axis = 1)

California_Confirmed_Sum
```

**Let's also create a dataframe `Daily_Confirmed_Sum` for daily confirmed counts.**

```python
Daily_C_Sum = pd.DataFrame(California_confirmed1.drop(['UID', 'Lat', 'Long'], axis = 1).sum(axis = 0)).rename(columns = {0 : 'Daily Confirmed Sum'})
Daily_Confirmed_Sum = Daily_C_Sum.T
Daily_Confirmed_Sum['Total'] = Daily_Confirmed_Sum.sum(axis = 1)
Daily_Confirmed_Sum
```

**The same treatment goes to the dataframe Death2, let's extract all the data for California and add a column `Death Sum` for the sum of the death cases for each county. Let's name the new dataframe `California_Death_Sum`.**

```python
California_death = Death2.iloc[Death2.index.get_level_values('Province/State') == 'California']
California_death1 = California_death[California_death['Lat'] != 0]
California_death1

Get_Death_Sum = pd.DataFrame(California_death1.drop(['UID', 'Lat', 'Long', 'Population'], axis = 1).sum(axis = 1)).rename(columns = {0:'Death Sum'})

California_Death_Sum = pd.concat([California_death1, Get_Death_Sum], axis = 1)

California_Death_Sum
```

**Let's also create a dataframe `Daily_Death_Sum` for daily death counts.**

```python
Daily_D_Sum = pd.DataFrame(California_death1.drop(['UID', 'Lat', 'Long', 'Population'], axis = 1).sum(axis = 0)).rename(columns = {0 : 'Daily Death Sum'})
Daily_Death_Sum = Daily_D_Sum.T
Daily_Death_Sum['Total'] = Daily_Death_Sum.sum(axis = 1)
Daily_Death_Sum
```
